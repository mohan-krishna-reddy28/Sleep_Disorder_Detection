```python
import numpy as np
import pandas as pd

dataset = pd.read_csv("../DATASET/sleep_prediction.csv")
dataset.head()
```

```python
dataset.isnull().sum()
```

```python
dataset["Sleep Disorder"] = dataset["Sleep Disorder"].fillna("No Disorder")
dataset["Sleep Disorder"].value_counts()
```

```python
dataset.isnull().sum()
```

```python
dataset.head(3)
```

```python
dataset["Blood Pressure"].value_counts()
```

```python
dataset[['Systolic', 'Diastolic']] = dataset["Blood Pressure"].str.split('/', expand=True).astype(int)
dataset.drop(columns=["Blood Pressure"], inplace=True)
```

```python
dataset.head()
```

```python
dataset['Gender'] = dataset['Gender'].map({'Male': 0, 'Female': 1})
```

```python
from sklearn.preprocessing import LabelEncoder

occ_encoder = LabelEncoder()
dataset["Occupation"] = occ_encoder.fit_transform(dataset['Occupation'])
for i, cls in enumerate(occ_encoder.classes_):
    print(i, "→", cls)
```

```python
dataset["Occupation"].value_counts()
```

```python
dataset = dataset.drop(columns=["Person ID"])
```

```python
bmi_encoder  = LabelEncoder()
target_encoder  = LabelEncoder()
dataset['BMI Category'] = bmi_encoder.fit_transform(dataset['BMI Category'])
dataset['Sleep Disorder'] = target_encoder.fit_transform(dataset['Sleep Disorder'])
for i, cls in enumerate(bmi_encoder.classes_):
    print(i, "→", cls)
for i, cls in enumerate(target_encoder.classes_):
    print(i, "→", cls)
```

```python
dataset.head()
```

```python
dataset["Sleep Disorder"].value_counts()
```

```python
from sklearn.model_selection import train_test_split
x = dataset.drop(columns=["Sleep Disorder"])
y = dataset["Sleep Disorder"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
x_train.shape,x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    max_depth=None,
    min_samples_split=2
)

rf.fit(x_train_scaled, y_train)

```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = rf.predict(x_test_scaled)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import joblib

joblib.dump(rf, "../BACKEND/RF_Sleep.pkl")
joblib.dump(scaler, "../BACKEND/scaler.pkl")
joblib.dump(occ_encoder, "../BACKEND/occ_encoder.pkl")
joblib.dump(bmi_encoder, "../BACKEND/bmi_encoder.pkl")  
joblib.dump(target_encoder, "../BACKEND/target_encoder.pkl")
```

```python
import numpy as np

# CNN expects (samples, height, width, channels)
X_train_cnn = x_train_scaled.reshape(x_train_scaled.shape[0], x_train_scaled.shape[1], 1, 1)
X_test_cnn = x_test_scaled.reshape(x_test_scaled.shape[0], x_test_scaled.shape[1], 1, 1)

X_train_cnn.shape, X_test_cnn.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, Flatten, Dense, Dropout

input_shape = (X_train_cnn.shape[1], 1, 1)

cnn = Sequential([
    Conv2D(32, (3,1), activation='relu', input_shape=input_shape),
    Conv2D(64, (3,1), activation='relu'),
    Flatten(),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(3, activation='softmax')
])

cnn.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

cnn.summary()
```

```python
history = cnn.fit(
    X_train_cnn, y_train,
    validation_split=0.2,
    epochs=25,
    batch_size=32,
    verbose=1
)

cnn.evaluate(X_test_cnn, y_test)
```

```python
cnn.save("../BACKEND/CNN_Sleep.h5")
```

```python
from tensorflow.keras.models import load_model

cnn = load_model("../BACKEND/CNN_Sleep.h5")
```

```python
import tensorflow as tf

converter = tf.lite.TFLiteConverter.from_keras_model(cnn)
converter.optimizations = [tf.lite.Optimize.DEFAULT]
quantized_model = converter.convert()

with open("../BACKEND/CNN_Sleep_quantized.tflite", "wb") as f:
    f.write(quantized_model)
```

```python
import numpy as np

X_train_lstm = x_train_scaled.reshape(x_train_scaled.shape[0], x_train_scaled.shape[1], 1)
X_test_lstm = x_test_scaled.reshape(x_test_scaled.shape[0], x_test_scaled.shape[1], 1)

X_train_lstm.shape, X_test_lstm.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

lstm = Sequential([
    LSTM(64, return_sequences=True, input_shape=(X_train_lstm.shape[1], 1)),
    Dropout(0.3),
    LSTM(32),
    Dropout(0.3),
    Dense(32, activation='relu'),
    Dense(3, activation='softmax')
])

lstm.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

lstm.summary()
```

```python
history_lstm = lstm.fit(
    X_train_lstm, y_train,
    validation_split=0.2,
    epochs=30,
    batch_size=32,
    verbose=1
)

lstm.evaluate(X_test_lstm, y_test)
```

```python
lstm.save("../BACKEND/LSTM_Sleep.h5")

converter = tf.lite.TFLiteConverter.from_keras_model(lstm)
converter.target_spec.supported_ops = [
    tf.lite.OpsSet.TFLITE_BUILTINS,
    tf.lite.OpsSet.SELECT_TF_OPS
]
converter._experimental_lower_tensor_list_ops = False
converter.optimizations = [tf.lite.Optimize.DEFAULT]

lstm_tflite = converter.convert()

with open("../BACKEND/LSTM_Sleep_quantized.tflite", "wb") as f:
    f.write(lstm_tflite)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import numpy as np

def save_confusion_matrix(y_true, y_pred, title, filename):
    cm = confusion_matrix(y_true, y_pred)
    fig, ax = plt.subplots()
    ax.imshow(cm)
    ax.set_title(title)
    ax.set_xlabel("Predicted")
    ax.set_ylabel("Actual")

    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(j, i, cm[i, j], ha="center", va="center")

    path = f"../BACKEND/{filename}"
    fig.savefig(path)
    plt.close(fig)
    return path

# RF confusion matrix
rf_pred = rf.predict(x_test_scaled)
rf_path = save_confusion_matrix(y_test, rf_pred, "RF Confusion Matrix", "RF_confusion.png")

# CNN confusion matrix
cnn_pred = np.argmax(cnn.predict(X_test_cnn), axis=1)
cnn_path = save_confusion_matrix(y_test, cnn_pred, "CNN Confusion Matrix", "CNN_confusion.png")

# LSTM confusion matrix
lstm_pred = np.argmax(lstm.predict(X_test_lstm), axis=1)
lstm_path = save_confusion_matrix(y_test, lstm_pred, "LSTM Confusion Matrix", "LSTM_confusion.png")

rf_path, cnn_path, lstm_path
```
